```python
# Keep only useful columns
data = data[['City', 'Date', 'PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']]

# Drop rows where ALL important pollutants are missing
data = data.dropna(subset=['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3'], how='all')

# Reset index
data = data.reset_index(drop=True)

# Check cleaned data
data.head()
```

```python
def get_sub_index(pollutant, concentration):
    # CPCB simplified breakpoints for demonstration
    if pollutant == 'NO2':
        if concentration <= 40:
            return concentration * 50 / 40
        elif concentration <= 80:
            return 50 + (concentration - 40) * 50 / 40
        else:
            return 100
    elif pollutant == 'SO2':
        if concentration <= 40:
            return concentration * 50 / 40
        elif concentration <= 80:
            return 50 + (concentration - 40) * 50 / 40
        else:
            return 100
    elif pollutant == 'CO':
        if concentration <= 1:
            return concentration * 50 / 1
        elif concentration <= 2:
            return 50 + (concentration - 1) * 50 / 1
        else:
            return 100
    elif pollutant == 'O3':
        if concentration <= 50:
            return concentration * 50 / 50
        elif concentration <= 100:
            return 50 + (concentration - 50) * 50 / 50
        else:
            return 100
    else:
        return None
```

```python
# Calculate sub-indices and AQI
def calculate_aqi(row):
    sub_indices = []
    for pollutant in ['NO2', 'SO2', 'CO', 'O3']:
        if not pd.isna(row[pollutant]):
            sub_index = get_sub_index(pollutant, row[pollutant])
            sub_indices.append(sub_index)
    return max(sub_indices) if sub_indices else None

data['Calculated_AQI'] = data.apply(calculate_aqi, axis=1)

data.head()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Convert 'Date' to datetime if not already
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')

# Extract Year
data['Year'] = data['Date'].dt.year

# Average AQI per year
yearly_aqi = data.groupby('Year')['Calculated_AQI'].mean().reset_index()

# Drop NA values if any
yearly_aqi.dropna(inplace=True)

# Prepare data for regression
X = yearly_aqi['Year'].values.reshape(-1,1)
y = yearly_aqi['Calculated_AQI'].values

# Train Linear Regression Model
model = LinearRegression()
model.fit(X, y)

# Predict for 2021 - 2025
future_years = np.array([[2021], [2022], [2023], [2024], [2025]])
future_aqi = model.predict(future_years)

# Show predictions
for year, pred in zip(future_years.flatten(), future_aqi):
    print(f"Predicted Average AQI for {year}: {pred:.2f}")

# Plot
plt.figure(figsize=(10,6))
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, model.predict(X), color='green', label='Regression Line')
plt.scatter(future_years, future_aqi, color='red', label='Predicted')
plt.legend()
plt.title('AQI Forecast (Linear Regression)')
plt.xlabel('Year')
plt.ylabel('Average AQI')
plt.grid(True)
plt.show()
```
